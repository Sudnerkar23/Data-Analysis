```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('top-youtubers-by-country-2024.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# Sort data by subscribers for better visualization
df = df.sort_values(by="TopYouTubers_TopChannelNumOfSubscribers_millions_2024", ascending=False)
```

```python
# Set style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df, 
    x="country", 
    y="TopYouTubers_NumberOfTopYouTubers_num_2024", 
    palette="coolwarm"
)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha="left")

# Add labels and title
plt.xlabel("Country")
plt.ylabel("Number of Top YouTubers")
plt.title("Top YouTubers by Country 2024 (min 55M subs)")

# Show the plot
plt.show()
```

```python
# Sort the dataframe by the number of top YouTubers for better visualization
df = df.sort_values(by="TopYouTubers_NumberOfTopYouTubers_num_2024", ascending=False)

# Set seaborn style and color palette
sns.set(style="darkgrid", context="talk", font_scale=1.2)
colors = sns.color_palette("viridis", len(df))  # Gradient color effect

# Create the bar plot
plt.figure(figsize=(14, 7))
bars = sns.barplot(
    data=df, 
    x="country", 
    y="TopYouTubers_NumberOfTopYouTubers_num_2024", 
    palette=colors
)

# Add labels on top of bars
for bar in bars.containers:
    bars.bar_label(bar, fmt='%d', label_type='edge', fontsize=12, padding=3, color='black')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha="right")

# Add labels and title
plt.xlabel("Country", fontsize=14, fontweight="bold")
plt.ylabel("Number of Top YouTubers", fontsize=14, fontweight="bold")
plt.title("Top YouTubers by Country (2024)", fontsize=16, fontweight="bold")

# Remove top and right spines for a clean look
sns.despine()

# Show the plot
plt.show()
```

```python
import plotly.express as px

# Create a choropleth map
fig = px.choropleth(
    df,
    locations="country",
    locationmode="country names",
    color="TopYouTubers_NumberOfTopYouTubers_num_2024",
    hover_name="country",
    color_continuous_scale="plasma",  # Choose aesthetic color theme
    title="Top YouTubers by Country (2024)",
)

# Update layout for better appearance
fig.update_layout(
    geo=dict(showcoastlines=True, projection_type="natural earth"),
    coloraxis_colorbar=dict(title="Number of Top YouTubers"),
    width=1000,  # Set width of the map
    height=500   # Set height of the map
)

# Show the plot
fig.show()
```
