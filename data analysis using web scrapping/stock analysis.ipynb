```python
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) -1
```

```python
ford['returns'] = (ford['Close']/ford['Close'].shift(1)) -1
```

# CUMMULATIVE RETURN

```python
tesla['Cummulative Returns'] = (1 + tesla['returns']).cumprod()
```

```python
tesla.head()
```

```python
ford['Cummulative Returns'] = (1 + ford['returns']).cumprod()
ford.head()
```

```python
gm['Cummulative Returns'] = (1 + gm['returns']).cumprod()
gm.head()
```

```python
tesla['Cummulative Returns'].plot( label='tesla',figsize=(13,7),kind='kde')
gm['Cummulative Returns'].plot( label="GM",kind='kde')
ford['Cummulative Returns'].plot( label='Ford',kind='kde')
plt.legend()
```
